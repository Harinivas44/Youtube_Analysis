```python
fig = plt.figure(figsize=(15, 10))
plt.subplot(2,2,1)
sns.histplot(data=df,x='view_count',kde=True)

plt.subplot(2,2,2)
sns.histplot(data=df,x='likes',kde=True)

plt.subplot(2,2,3)
sns.histplot(data=df,x='dislikes',kde=True)

plt.subplot(2,2,4)
sns.histplot(data=df,x='comment_count',kde=True)

plt.show()
```

```python
cat=df['category'].value_counts()
```

```python
sns.barplot(x=cat.index,y=cat.values)
plt.xticks(rotation=90)
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.title('Categories on which youtube videos are Uploaded')
plt.show()
```

```python
correlation=df.corr(numeric_only=True).iloc[1:5,1:5]
correlation
```

```python
sns.heatmap(correlation,annot=True)
```
